```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataset sucio

df_original = pd.read_csv("../data_Proyecto/ecommerce_customer_behavior_dataset_Dirty.csv")
df = df_original.copy()

print("Shape inicial:", df.shape)
df.head()
```

```python
# Datos faltantes

faltantes = df.isna().sum().sort_values(ascending=False)
porcentaje = (df.isna().mean()*100).round(2)

print("\n--- Valores faltantes por columna ---\n")
print(pd.DataFrame({"faltantes": faltantes, "porcentaje_%": porcentaje}))

# Eliminamos SOLO filas completamente vacías (si existieran)
df.dropna(how="all", inplace=True)

print("\nShape después de eliminar filas vacías:", df.shape)
```

```python
# Columnas

columnas_categoricas = df.select_dtypes(include="object").columns
columnas_numericas   = df.select_dtypes(exclude="object").columns

print("\nColumnas categóricas:", list(columnas_categoricas))
print("Columnas numéricas:", list(columnas_numericas))
```

```python
# Filas repetidas

duplicados = df[df.duplicated(keep=False)]
print("\n--- Filas duplicadas detectadas ---")
print(duplicados.head())

print(f"\nRegistros duplicados: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print("Shape después de eliminar duplicados:", df.shape)
```

```python
# Errores tipográficos y uniformización

# Convertir todas las categorías a minúsculas para uniformar
for col in columnas_categoricas:
    df[col] = df[col].astype(str).str.lower()

for col in columnas_categoricas:
    df[col] = df[col].str.strip()


# Revisión
print("\n--- Categorías unificadas (primeras columnas) ---\n")
for col in columnas_categoricas[:5]:
    print(col, df[col].unique(), "\n")
```

```python
# Guardar dataset limpio

df.to_csv("ecommerce_customer_behavior_Clean.csv", index=False)
print("\nArchivo guardado como ecommerce_customer_behavior_Clean.csv")
print("Shape final:", df.shape)
```
