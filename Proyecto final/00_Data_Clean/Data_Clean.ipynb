# Limpieza y enriquecimiento del dataset de e-commerce

```python
import pandas as pd
from pathlib import Path

raw_path = Path("../ecommerce_customer_behavior_dataset.csv")
df = pd.read_csv(raw_path)
print("Forma inicial:", df.shape)
df.head()
```

```python
# Checar tipos y valores nulos
print(df.dtypes)
print("
Nulos por columna:
", df.isnull().sum())
```

```python
# Enriquecimiento simple de fecha y guardado del dataset limpio
df["Date"] = pd.to_datetime(df["Date"])
df["order_month"] = df["Date"].dt.month
df["order_dayofweek"] = df["Date"].dt.dayofweek

clean_df = df.drop_duplicates().copy()
out_path = Path("Data_Set.csv")
clean_df.to_csv(out_path, index=False)
print("Dataset limpio guardado en", out_path.resolve())
clean_df.head()
```
