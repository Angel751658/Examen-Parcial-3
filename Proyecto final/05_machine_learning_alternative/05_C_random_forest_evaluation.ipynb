```python
import pandas as pd
import numpy as np
import joblib
from pathlib import Path
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
prep_dir = Path("../01_preprocessing_results/preprocessing")
X_test = pd.read_csv(prep_dir / "T_test_final_objetivo.csv")
objetivo = "objetivo"
y_test = X_test[objetivo]
X_test = X_test.drop(columns=[objetivo])

rf = joblib.load("modelo_random_forest.pkl")

preds = rf.predict(X_test)
```

```python
# Metricas
mse = mean_squared_error(y_test, preds)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, preds)
r2 = r2_score(y_test, preds)

pd.DataFrame({
    "mse": [mse], "rmse": [rmse], "mae": [mae], "r2": [r2]
}).to_csv("evaluacion_random_forest.csv", index=False)

print("MSE:", mse)
print("MAE:", mae)
print("R²:", r2)
print("RMSE:", rmse)
```

```python
# Importancias
importancias = pd.Series(rf.feature_importances_, index=X_test.columns)
importancias.sort_values(ascending=False).to_csv("rf_feature_importances.csv")

importancias
```
