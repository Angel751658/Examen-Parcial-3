```python
import pandas as pd
import numpy as np
```

```python
# Archivo subido en esta sesión
precios_reales_path = "./data/Saratoga_Precios_Reales.txt"

precios_reales = pd.read_csv(
    precios_reales_path,
    header=None,
    names=["Precio_Real"]
)

precios_reales
```

```python
pred_path = "./04_regularization/Regresion_ganadora_nuevos_predicciones.csv"

predicciones = pd.read_csv(pred_path)

# Asegurar que existe yhat
if "yhat" not in predicciones.columns:
    raise ValueError("ERROR: El archivo no contiene columna 'yhat'.")

predicciones.head()
```

```python
n = min(len(precios_reales), len(predicciones))

df_comp = pd.DataFrame({
    "Precio_Real": precios_reales["Precio_Real"][:n],
    "Prediccion": predicciones["yhat"][:n]
})

df_comp["Error_Absoluto"] = abs(df_comp["Precio_Real"] - df_comp["Prediccion"])
df_comp["Error_Relativo_%"] = 100 * df_comp["Error_Absoluto"] / df_comp["Precio_Real"]

df_comp
```

```python
MAE = df_comp["Error_Absoluto"].mean()
RMSE = np.sqrt((df_comp["Error_Absoluto"]**2).mean())
MAPE = df_comp["Error_Relativo_%"].mean()

print("=== MÉTRICAS FINALES DEL MODELO ===")
print(f"MAE  : {MAE:,.2f}")
print(f"RMSE : {RMSE:,.2f}")
print(f"MAPE : {MAPE:.2f}%")
```

```python
df_comp.to_csv("Comparacion_Final_Modelo_vs_Reales.csv", index=False)
print("Archivo guardado: Comparacion_Final_Modelo_vs_Reales.csv")
```

```python
import joblib
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Cargar modelo ganador
modelo = joblib.load("./04_regularization/reg_lin_ganador/reg_lin_ganador_bundle/modelo_elasticnet.pkl")

# Cargar datos
Train = pd.read_csv("./01_preprocessing_results/preprocessing/T_train_final_objetivo.csv")
Test = pd.read_csv("./01_preprocessing_results/preprocessing/T_test_final_objetivo.csv")

X_train = Train.iloc[:, :-1]
y_train = Train.iloc[:, -1]

X_test = Test.iloc[:, :-1]
y_test = Test.iloc[:, -1]

# Predicciones
yhat_train = modelo.predict(X_train)
yhat_test = modelo.predict(X_test)

# Métricas Train
MAE_train = mean_absolute_error(y_train, yhat_train)
RMSE_train = np.sqrt(mean_squared_error(y_train, yhat_train))
R2_train = r2_score(y_train, yhat_train)

# Métricas Test
MAE_test = mean_absolute_error(y_test, yhat_test)
RMSE_test = np.sqrt(mean_squared_error(y_test, yhat_test))
R2_test = r2_score(y_test, yhat_test)

print("=== MÉTRICAS ENTRENAMIENTO ===")
print(f"MAE_train  = {MAE_train:,.2f}")
print(f"RMSE_train = {RMSE_train:,.2f}")
print(f"R2_train   = {R2_train:.4f}")

print("\n=== MÉTRICAS PRUEBA ===")
print(f"MAE_test  = {MAE_test:,.2f}")
print(f"RMSE_test = {RMSE_test:,.2f}")
print(f"R2_test   = {R2_test:.4f}")
```
